Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.


```python
# Re-run this cell 
import pandas as pd
import numpy as np

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...

# Calculate total SAT score
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']

# Filter schools with math results of at least 80% of the maximum possible score of 800
best_math_schools = schools[schools['average_math'] >= 0.8 * 800][['school_name', 'average_math']].sort_values(by='average_math', ascending=False)

# Sort schools based on the combined SAT scores
top_10_schools = schools[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False).head(10)


# Calculate standard deviation of total SAT score for each borough
borough_std_dev = schools.groupby('borough')['total_SAT'].std().reset_index()
largest_std_dev = borough_std_dev.loc[borough_std_dev['total_SAT'].idxmax()]
largest_borough = largest_std_dev['borough']

# Calculate the number of schools in the borough with the largest standard deviation
num_schools = schools['borough'].value_counts()[largest_borough]

# Calculate the average SAT score for the borough with the largest standard deviation
average_SAT = schools.groupby('borough')['total_SAT'].mean().loc[largest_borough]

# Create the DataFrame
largest_std_dev = pd.DataFrame({
    'borough': [largest_borough],
    'num_schools': [num_schools],
    'average_SAT': [average_SAT],
    'total_SAT': [largest_std_dev['total_SAT']]
})

# Ensure the columns are numerical
largest_std_dev['total_SAT'] = pd.to_numeric(largest_std_dev['total_SAT'], errors='coerce')
largest_std_dev['num_schools'] = pd.to_numeric(largest_std_dev['num_schools'], errors='coerce')

# Rename the column
largest_std_dev.rename(columns={'total_SAT': 'std_SAT'}, inplace=True)

# Set the index to the borough
largest_std_dev.set_index('borough', inplace=True)


# Round numeric values in largest_std_dev to two decimal places
largest_std_dev = largest_std_dev.round(2)


print(f"Best Math Schools: \n {best_math_schools}")
print(f"\nTop 10 Performing Schools based on Combined SAT Scores: \n {top_10_schools}")
print(f"\nBorough with the Largest Standard Deviation in Combined SAT Score: \n {largest_std_dev}")

```
